## Now we are ready to clean out dataset

```python
# Lets import the neccesarry libraries 
import pandas as pd 

# Lets create out dataframe
cost_df = pd.read_csv('health-insurance-marketplace/BenefitsCostSharing.csv')
cost_df
```

```python
# Lets see the size and shape of our data set
import os 

cost_df_size = os.path.getsize('health-insurance-marketplace/BenefitsCostSharing.csv')/1e6
cost_df_shape = cost_df.shape 
print("The size of the data set as a file is " , cost_df_size , " mb.")
print("The rows and columns ares: ", cost_df_shape)
```

```python
# Lets search for NaN or null values. 

# If we choose a column we can filter that column base in criteria 
# in the next bracket just like this: [column][criteria], the criteria returns a bool value

# Lets check if in the BenefitName column there are null values
cost_df['BenefitName'][cost_df['BenefitName'].isna()]
```

```python
# Lets check if in the CoinsInnTier1 there are some null values. 
cost_df['CoinsInnTier1'][cost_df['CoinsInnTier1'].isna()]

#Lets check how many nun values are
cost_df['CoinsInnTier1'].isnull().sum()
```

```python
# Lets check if in the CoinsInnTier1 there are some null values. 
cost_df['CoinsInnTier2'][cost_df['CoinsInnTier2'].isna()]

#Lets check how many nun values are
cost_df['CoinsInnTier2'].isnull().sum()
```

```python
# Lets work with CoinsInnTier1 and CoinsInnTier2 
focus_cols = ['CoinsInnTier1', 'CoinsInnTier2']

# lets analize this line of code part by part: 
# 1.  cost_df[], here we are going to filter entire dataset.
# 2.  Inside cost_df[] we have to put the columns with the criteria
# 3.  cost_df[focus_cols].isna() this outputs boolean values, with .all() 
        # all() method checs if all elements are true in an axis.
        # 0 index or 1 columns

# Example in entire data set if whe want only entries in which
# CoinsInnTier1 and 2 are not null we should: 
cost_df[cost_df[focus_cols].notna().all(1)]

# if we wnat only entries in whinch Coins 1 and 2 are null

# this one alse tell us where both are null apparently most of columns are null. 
cost_df[cost_df[focus_cols].isna().all(1)] 

# Note that this is a criteria, we used it to filter
cost_df[focus_cols].notna().all(1)

# like this: 
cost_df[focus_cols][cost_df[focus_cols].notna().all(1)]

# So in conclusion to filter entirer data set cost_df[criteria]
# to filter columns or column cost_df[colums][criteria]


```
